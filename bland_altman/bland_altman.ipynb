```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def bland_altman_plot(data1, data2):
    data1 = np.asarray(data1)
    data2 = np.asarray(data2)
    means = np.mean([data1, data2], axis=0)
    differences = data1 - data2
    mean_of_differences = np.mean(differences)
    std_dev = np.std(differences, axis=0)

    plt.scatter(means, differences)
    plt.axhline(mean_of_differences, color='gray', linestyle='--')
    plt.axhline(mean_of_differences + 1.96 * std_dev, color='gray', linestyle='--')
    plt.axhline(mean_of_differences - 1.96 * std_dev, color='gray', linestyle='--')
    plt.xlabel("Average of 2 measures")
    plt.ylabel("Difference between 2 measures")
    plt.title("Bland-Altman Plot")
    plt.show()
```

```python
df = pd.DataFrame({'A': [5, 5, 5, 6, 6, 7, 7, 7, 8, 8, 9,
                         10, 11, 13, 14, 14, 15, 18, 22, 25],
                   'B': [4, 4, 5, 5, 5, 7, 8, 6, 9, 7, 7, 11,
                         13, 13, 12, 13, 14, 19, 19, 24]})

bland_altman_plot(data1=df["A"], data2=df["B"])
```
