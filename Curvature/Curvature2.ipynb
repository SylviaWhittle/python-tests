```python
import autograd.numpy as np
import math
from autograd import grad

# Plot function thing
def f(x):
    #return 1/(1 + np.exp(-x))
    return np.sin(x/2)

# 1st Derivative
f1 = grad(f)  
# 2nd Derivative
f2 = grad(f1) 

# Calculate curvature function
def curvature(x):
    return abs(f2(x))*(1 + f1(x)**2)**-1.5
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 20)
plt.plot(x, f(x), 'ko')
plt.title("Data")
plt.show()

y = [curvature(t) for t in x]
plt.plot(x, y, 'ko')
plt.title("Curvature")
```
