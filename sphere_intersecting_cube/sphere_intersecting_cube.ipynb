```python
import random
import time

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
size = 256

x, y, z = np.mgrid[0:size, 0:size, 0:size]
points = np.vstack((z.ravel(), y.ravel(), x.ravel())).T
# print(points)
print(points.shape)
print(points[0], points[1], points[2])

# Create sphere
sphere_radius = random.randrange(int(size / 5), int(size / 2))
print(sphere_radius)
sphere_centre = np.array(
    [
        random.randrange(sphere_radius, size - sphere_radius),
        random.randrange(sphere_radius, size - sphere_radius),
        random.randrange(size, size + sphere_radius - 1),
    ]
)

print(f"sphere centre: {sphere_centre} sphere radius: {sphere_radius}")

# Determine which points are inside the sphere
sphere_mask = np.linalg.norm(points - sphere_centre, axis=1) < sphere_radius

print(f"sphere mask shape: {sphere_mask.shape}, unique: {np.unique(sphere_mask)}")


# Make animation of the frames
fig = plt.figure(figsize=(5, 5))
im = plt.imshow(np.random.random((256, 256)), interpolation="none")


def animate(i):
    box_slice = np.array(sphere_mask[i * size**2 : (i + 1) * size**2]).reshape(size, size)
    im.set_array(box_slice)
    plt.title(f"depth: {i}")
    return [im]


n_seconds = 2
total_seconds = 4
fps = size / total_seconds
anim = animation.FuncAnimation(
    fig,
    animate,
    frames=size,
    interval=1000 / fps,
)


anim.save("sphere_slices.mp4")


# fig = plt.figure(figsize=(10, 7))
# ax = plt.axes(projection="3d")
# ax.scatter3D(points[:, 0], points[:, 1], points[:, 2], color="green")
# plt.title("simple 3D scatter plot")
# plt.show()

# sphere_points = points[sphere_mask]

# fig = plt.figure(figsize=(10, 7))
# ax = plt.axes(projection="3d")
# ax.scatter3D(sphere_points[:, 0], sphere_points[:, 1], sphere_points[:, 2], color="orange")
# plt.title("simple 3D scatter plot")
# plt.show()
```
